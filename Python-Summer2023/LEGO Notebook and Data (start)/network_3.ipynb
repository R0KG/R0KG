```r
install.packages("ggplot2")
# Load the ggplot2 library
library(ggplot2)

# Define the formula as an R function
evaluate_formula <- function(x) {
  y <- x * (1/x) * ((x-1)/x)^(x-1)
  return(y)
}

# Create a data frame with x and y values
data <- data.frame(x = 1:100000, y = sapply(1:100000, evaluate_formula))

# Create a scatterplot
ggplot(data, aes(x = x, y = y)) +
  geom_point() +
  labs(x = "x values", y = "Formula Output") +
  ggtitle("Formula Evaluation from 1 to 100000")
```
